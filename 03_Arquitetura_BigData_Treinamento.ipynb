# Resumo sobre alguns aplicativos

```python
### 1) ZooKeeper: ZooKeeper  um servio de coordenao distribuda para gerenciar grandes conjuntos de Clusters.

### 2) Oozie: Apache Oozie  um sistema de agendamento de WorkFlow, usado para gerenciar principalmente os Jobs de MapReduce.

### 3) Pig: Apache Pig,  uma linguagem de procedimentos de alto nvel para consultar grandes conjuntos de dados semiestruturados usando Hadoop e a Plataforma MapReduce

### 4) Sqoop: Apache Sqoop,  um projeto do ecossistema Hadoop, cuja responsabilidade  importar e exportar dados do banco de dados de dados relacionais.

### 5) Spark: Apache Spark,  uma ferramenta Big Data para o processamento de grandes conjuntos de dados. Foi desenvolvido para substituir o MapReduce, pois processa 100x mais rpido que o MapReduce.

### 6) HBase: Apache Hbase,  um banco de Dados no relacionais, projetado para trabalhar com grande conjunto de dados (Big Data).  o banco de dados oficial do hadoop.

### 7) Flume: Apache Flume,  um servio que permite enviar dados diretamente para o HDFS.  um servio que funciona em ambiente dstribuido (em cluster) para coletar, agregar e mover grandes quantidades de dados de forma eficiente.

### 8) Mahout: Apache Mahout,  dedicado a Machine Learning  Data Science. Ele permite a utilizao dos principais algoritmos de clustering, testes de regresso e modelagem estatstica.

### 9) Kafka: Apache Kafka,  foi desenvolvido pelo Linkedin e liberado como projeto Open-source em 2011. O Apache Kafka  um sistema para gerenciamento de fluxo de dados em tempo real, gerados a partir de websites, aplicaes e sensores.

### 10) Ambari: Apache Ambari tem como objetivo tornar o gerenciamento do Hadoop mais simples. O Ambari fornece uma interface de usurio da Web de gerenciamento do Hadoop intuitiva e fcil de usar.

```
