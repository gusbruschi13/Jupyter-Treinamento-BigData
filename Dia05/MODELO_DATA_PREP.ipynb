LER TABELA DE METADADOS

```python
import pandas as pd

metadados = pd.read_csv('metadados_app_train.csv',
                        sep=';',
                        encoding='utf-8')
metadados
```

Selecionando somente as variÃ¡veis explicativas (Ordinal+Interval)

```python
# Filtrando Data Frame 
vars_numericas_df = metadados[((metadados.Level  == 'ordinal')|(metadados.Level == 'interval')) & (metadados.Role == 'input')]
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
vars_numericas_df
```

```python
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
lista_vars_numericas = list(vars_numericas_df['Features'])
lista_vars_numericas
```

```python
# Vamos montar um data frame com as variÃ¡veis que foram filtradas (Note que a tabela ja tem indice)
df_app_train_001 = df_app_train[lista_vars_numericas]
df_app_train_001.head()
```

```python
len(df_app_train_001)
```

TRATAMENTO DOS MISSINGS

```python
# Primeiro vamos verificar a quantidade de missings em todas variÃ¡veis
df_missing_001 = df_app_train_001.isnull().sum()
df_missing_001
```

```python
# Salvando quantidade de linhas da tabela em variÃ¡vel
qt_rows = df_app_train_001.shape[0]

# Gerando data frame com quantidade de missings por variavel
df_pct_missing = pd.DataFrame(df_missing_001,columns=['qt_missing'])
df_pct_missing = pd.DataFrame(df_missing_001,columns=['qt_missing'])
df_pct_missing['Features'] = df_pct_missing.index
df_pct_missing['pc_miss'] = (100*df_pct_missing['qt_missing'].divide(qt_rows)).astype(int)
df_pct_missing['qt_rows'] = qt_rows
df_pct_missing.reset_index(drop = True, inplace = True)

df_pct_missing
```

```python
# Aplicar o tratamento dos missings substituindo por mÃ©dia 
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.fillna.html

df_app_train_002 = df_app_train_001.fillna(df_app_train_001[lista_vars_numericas].mean())

print('Verificando se ainda tem missings apÃ³s tratamento:')
df_app_train_002.isnull().sum()
```

```python
df_app_train_002.head(20)
```

```python
from sklearn.preprocessing import StandardScaler

# Classe responÃ¡vel pela normalizaÃ§Ã£o
scaler = StandardScaler()

# Convertendo todas variÃ¡veis para tipo float (necessÃ¡rio para normalizaÃ§Ã£o)
df_app_train_003 = df_app_train_002.astype(float)

scaled_features = scaler.fit_transform(df_app_train_003[lista_vars_numericas])
df_app_train_004 = pd.DataFrame(scaled_features, columns=lista_vars_numericas)

df_app_train_004.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Classe responÃ¡vel pela normalizaÃ§Ã£o
scaler_amplitude = MinMaxScaler(feature_range=(0, 1))

scaled_features = scaler_amplitude.fit_transform(df_app_train_003[lista_vars_numericas])
df_app_train_005 = pd.DataFrame(scaled_features, columns=lista_vars_numericas)

df_app_train_005.head(10)
```

Selecionando somente as variÃ¡veis do tipo nominal

```python
# Filtrando Data Frame 
vars_char_df = metadados[(metadados.Level  == 'nominal') & (metadados.Role == 'input')]
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
vars_char_df
```

```python
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
lista_vars_char = list(vars_char_df['Features'])
lista_vars_char
```

```python
df_app_train_006 = df_app_train[lista_vars_char]
df_app_train_006.head()
```

```python
df_app_train_006_missing = df_app_train_006.isnull().sum()
df_app_train_006_missing
```

```python
df_app_train_007 = df_app_train_006.fillna('MISS')
df_app_train_007.head()
```

```python
df_app_train_007_missing = df_app_train_007.isnull().sum()
df_app_train_007_missing
```

```python
# Filtrando Data Frame 
vars_to_dummif_df = metadados[(metadados.Level  == 'nominal') & (metadados.Role == 'input')& (metadados.Cardinality <= 10)]
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
vars_to_dummif_df
```

```python
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
lista_vars_dummif = list(vars_to_dummif_df['Features'])
lista_vars_dummif
```

```python
df_app_train_008 = df_app_train_007[lista_vars_dummif]
df_app_train_008.head()
```

```python
# gerando variÃ¡veis dummies
df_app_train_009 = pd.get_dummies(df_app_train_008, 
                      columns=lista_vars_dummif,
                      drop_first=False, 
                      prefix = lista_vars_dummif,
                      prefix_sep='_')
df_app_train_009.head()
```

```python
# Filtrando Data Frame (Retiramos a variÃ¡vel Name)
vars_to_le_df = metadados[(metadados.Level  == 'nominal') & 
                          (metadados.Role == 'input') & 
                          (metadados.Cardinality > 10) & 
                          (metadados.Features != 'Name')]

# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
vars_to_le_df
```

```python
# Selecionando nomes das variÃ¡veis a partir do data frame filtrado
lista_vars_le = list(vars_to_le_df['Features'])
lista_vars_le
```

```python
df_app_train_010 = df_app_train_007[lista_vars_le]
df_app_train_010.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

le_occ_type = le.fit_transform(df_app_train_010['OCCUPATION_TYPE'])
le_le_occ_type_df = pd.DataFrame(le_occ_type, columns=['LE_OCCUPATION_TYPE'])

le_org_type = le.fit_transform(df_app_train_010['ORGANIZATION_TYPE'])
le_org_type_df = pd.DataFrame(le_org_type, columns=['LE_ORGANIZATION_TYPE'])

df_app_train_011 = pd.merge(le_le_occ_type_df,le_org_type_df, left_index=True, right_index=True)

df_app_train_011.head()
```

```python
df4_9 = pd.merge(df_app_train_004,df_app_train_009, left_index=True, right_index=True)
df_ABT_expl = pd.merge(df4_9,df_app_train_011, left_index=True, right_index=True)
df_ABT_expl.head()
```

Vamos trazer o Target para gerar a ABT final (pronta para o treinamento dos modelos de ML)

```python
ABT_df = pd.merge(df_ABT_expl,df_app_train[['target']], left_index=True, right_index=True)
ABT_df.head()
```
